```python
d1 = pd.read_csv('garden_2.csv')
d2 = pd.read_csv('garden_3.csv')
d3 = pd.read_csv('garden_99_2032.csv')
d4 = pd.read_csv('garden_without_mixup_1.csv')
d5 = pd.read_csv('garden_without_mixup_2.csv')
d6 = pd.read_csv('garden_without_mixup_3.csv')
d7 = pd.read_csv('garden_without_mixup_4.csv')
d8 = pd.read_csv('garden_without_mixup_5_rep.csv')
d9 = pd.read_csv('garden_without_mixup_5_rep.csv')
d10 = pd.read_csv('garden__mixup_heavy_augment_2.csv')
```

```python
d1.head()
```

```python
ensemble = pd.DataFrame({'image_id':d1.image_id,
                         'd1':d1.category,
                         'd2':d2.category,
                         'd3':d3.category,
                         'd4':d4.category,
                         'd5':d5.category,
                         'd6':d6.category,
                         'd7':d7.category,
                         'd8':d8.category,
                         'd9':d9.category,
                         'd10':d10.category
                        })
```

```python
ensemble['category'] = 1
```

```python
ensemble['category'] = ensemble.mode(axis=1)
```

```python
ensemble.dtypes
```

```python
ensemble['category'] = ensemble['category'].astype('int64')
```
